```python
# Date preprocessing

import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 

dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:, 1:2]
y = dataset.iloc[:, -1]

# Let us not split this one

# Linear Regression
from sklearn.linear_model import LinearRegression
regressor_linear = LinearRegression()
regressor_linear.fit(X, y)
# we did this only for comparison purpose
```

```python
from sklearn.preprocessing import PolynomialFeatures

# we first converrt our X to polynomial, then use the LinearRegression model, that's how polynomial regression is done
poly_converter = PolynomialFeatures(degree=4)
X_poly = poly_converter.fit_transform(X)
poly_converter.fit(X_poly, y)

poly_regressor = LinearRegression()
poly_regressor.fit(X_poly, y)
```

```python
# Let us visualize them

# Linear Regression
plt.scatter(X, y, color="red")
plt.plot(X, regressor_linear.predict(X), color="blue")
plt.title("Linear Regression: Level Vs Salary")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
# Polynomial Regression



plt.scatter(X, y, color="red")
plt.plot(X, poly_regressor.predict(poly_converter.fit_transform(X)), color="blue")
plt.title("Polynomial Regression: Level Vs Salary")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
# predict for specific employee
regressor_linear.predict([[6.5]])
```

```python
poly_regressor.predict(poly_converter.fit_transform([[6.5]]))
```
